```python
from matplotlib import pyplot as plt
import numpy as np
import imageio
from skimage import color
```

```python
ruta = 'images/gray-wolf.jpg'
imgIn = imageio.v2.imread(ruta)
```

```python
print(imgIn.shape, imgIn.dtype)
```

# Escala de grises

```python
imgGray = color.rgb2gray(imgIn)
print(imgGray.shape, imgGray.dtype)
```

```python
plt.imshow(imgGray, cmap='gray')
```

```python
imgSection = imgGray[110:440,215:525]
plt.imshow(imgSection, cmap='gray')
```

```python
imgModificada = imgGray.copy()
imgModificada[imgModificada < 0.25] = 0.0
plt.imshow(imgModificada, cmap = 'gray')
```

```python
plt.hist(imgGray)
```

```python
plt.hist(imgGray.flatten())
```

```python
plt.hist(imgModificada.flatten())
```

```python
imageio.imwrite('graywolf-modified.jpg', (imgModificada*225).astype('uint8'))
```
