```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import metrics
```

```python
data = sns.load_dataset('penguins')
```

```python
train, test = train_test_split(data, test_size = 0.4, stratify = data['species'], random_state = 17)
```

```python
fn = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
cn = ['gentoo', 'chinstrap', 'adelie']
```

```python
x_train = train[fn]
y_train = train['species']
```

```python
x_test = test[fn]
y_test = test['species']
```

```python
mod_dt = DecisionTreeClassifier(max_depth = 3, random_state = 112)

mod_dt.fit(x_train, y_train)

prediction = mod_dt.predict(x_test)
```

```python
mod_dt.feature_importances_
```

```python
plt.figure(figsize = (20, 12))
plot_tree(mod_dt, feature_names = fn, class_names = cn, filled = True);
```

```python
efficiency = metrics.accuracy_score(y_test, prediction)
efficiency
```

```python
compare = pd.DataFrame(x_test)
compare['species'] = y_test
compare['prediction'] = prediction
compare
```
