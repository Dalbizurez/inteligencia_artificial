```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from io import StringIO
```

```python
url = 'https://es.wikipedia.org/wiki/Organizaci%C3%B3n_territorial_de_Guatemala'
header = {'User-Agent': 'NoSoyBot/8.0 (contacto:danielalbizurez@gmail.com)'}
```

```python
response = requests.get(url, headers = header)
```

```python
print(f'Status Code {url}: {response.status_code}')
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
tabla = soup.find('table', {'class':'wikitable'})
```

```python
tabla = pd.read_html(StringIO(str(tabla)))
```

```python
tabla
```

```python
departamento = tabla[1]
```

```python
departamento.columns = ['Departamento', 'Cabecera', 'Superficie', 'Población']
departamento
```

```python
region = soup.find_all('table')[2]
region = pd.read_html(StringIO(str(region)))[0]
region
```

```python
segeplan = soup.find_all('table')[2]
segeplan = pd.read_html(StringIO(str(segeplan)))[0]
segeplan
```

```python
subregion = soup.find_all('table')[2]
subregion = pd.read_html(StringIO(str(subregion)))[0]
subregion
```

```python
departamento['Superficie'] = list(int(x.replace('\xa0', '')) for x in departamento['Superficie'])
```

```python
departamento['Población'] = list(int(x.replace('\xa0', '')) for x in departamento['Población'])
```
