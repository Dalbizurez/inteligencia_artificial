Modelos que se encargan de identificar y clasificar datos

Una de las formas es a través de los arboles de decisión

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importar el separador de muestras para entrenamiento y pruebas
from sklearn.model_selection import train_test_split
# Importar el clasificador: Árbol de decisión
from sklearn.tree import DecisionTreeClassifier, plot_tree
# Importar las métricas para medir la eficicienia del modelo
from sklearn import metrics
```

```python
# Cargar el dataset
data = sns.load_dataset('iris')
```

```python
# Separación de datos en Entrenamiento-Prueba
train, test = train_test_split(data, test_size = 0.4, stratify = data['species'], random_state = 26)
```

```python
fn = ['petal_length', 'petal_width']
cn = ['setosa', 'versicolor', 'virginica']
```

```python
X_train = train[fn]
y_train = train['species']
```

```python
# Conjunto de pruebas siguiendo el modelo y =mX + b
X_test = test[fn]
y_test = test['species']
```

```python
# Generar el modelo en base al clasificador
# Arbol de decision
mod_dt = DecisionTreeClassifier(max_depth = 3, random_state = 112)

# Ajustar el modelo
mod_dt.fit(X_train, y_train)

# Probar el modelo
prediccion =  mod_dt.predict(X_test)
```

```python
mod_dt.feature_importances_
```

```python
# Visualizar las reglas de clasificacion
# tupla(ancho, altura)
plt.figure(figsize=(10, 8))
plot_tree(mod_dt, feature_names = fn, class_names=cn, filled=True);
```

```python
# Eficiencia del modelo
eficiencia = metrics.accuracy_score(y_test, prediccion)
eficiencia
```

```python
coso = pd.DataFrame(X_test)
coso['species'] = y_test
coso['prediccion'] = prediccion
coso
```
