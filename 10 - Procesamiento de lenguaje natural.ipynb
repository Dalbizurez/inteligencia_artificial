```python
from transformers import BertTokenizer
```

```python
import torch
torch.cuda.empty_cache()
```

```python
tokenizer = BertTokenizer.from_pretrained('google-bert/bert-base-uncased')
```

```python
texts_en = [
    "Artificial intelligence is transforming the world.",
    "Python is a versatile and powerful programming language.",
    "Data science combines statistics and computer science.",
    "Machine learning enables systems to learn from data.",
    "Cybersecurity is essential in the digital age.",
    "Cloud computing provides scalable resources online.",
    "Open-source software encourages collaboration.",
    "Automation increases efficiency and productivity.",
    "Big data requires advanced analytics tools.",
    "Continuous learning is key in technology careers."
]

texts_es = [
    "La inteligencia artificial está transformando el mundo.",
    "Python es un lenguaje de programación versátil y potente.",
    "La ciencia de datos combina estadística e informática.",
    "El aprendizaje automático permite a los sistemas aprender de los datos.",
    "La ciberseguridad es esencial en la era digital.",
    "La computación en la nube ofrece recursos escalables en línea.",
    "El software de código abierto fomenta la colaboración.",
    "La automatización aumenta la eficiencia y la productividad.",
    "El big data requiere herramientas avanzadas de análisis.",
    "El aprendizaje continuo es clave en las carreras tecnológicas."
]
```

```python
tokens_es = tokenizer.tokenize(texts_es)
tokens_en = tokenizer.tokenize(texts_en)
```

```python
tokens_id_en = tokenizer.convert_tokens_to_ids(tokens_en)
tokens_id_en
```

```python
texto_en_decode = tokenizer.decode(tokens_id_en)
texto_en_decode
```

```python
for i, oracion in enumerate(texts_en):
    # encode añade automáticamente [CLS] y [SEP]
    codificado = tokenizer._encode_plus(
        oracion,
        add_special_tokens = True,
        max_length = 20,
        padding = 'max_length',
        truncation = True,
        return_tensors = 'pt',
        return_attention_mask = True
    )
    print(f'\noracion{i+1}: {oracion}')
    print(f'Ids: {codificado['input_ids'][0].tolist()}')
    print(f'Mascara de atencion {codificado['attention_mask'][0].tolist()}')
```

```python
batch_encode = tokenizer(
    texts_en,
    max_length = 20,
    padding = 'max_length',
    truncation = True,
    return_tensors = 'pt'
)
```
